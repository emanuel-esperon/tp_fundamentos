```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

rawDataset = pd.read_csv("Covers.csv")

copi = rawDataset.copy()

def tiempo_a_segundos(t):
    return int(t[0])*60 + int(t[2])*10 + int(t[3])

copi['Duration'] = copi['Duration'].apply(tiempo_a_segundos)

```

```python

numeros = copi.select_dtypes(include=['number'])
correlationMatrix = numeros.corr()
plt.figure(figsize=(16,12))
sns.heatmap(correlationMatrix, vmin=-1.0, vmax=1.0, center=0.0, annot=True, cmap= 'coolwarm')
plt.savefig('Heatmap_corr.png')
plt.show()
```
